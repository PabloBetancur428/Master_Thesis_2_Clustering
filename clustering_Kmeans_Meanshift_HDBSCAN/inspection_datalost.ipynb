```python
import pandas as pd

# Example CSV (update with your actual file)
input_csv = "/home/jbetancur/Desktop/codes/clustering/clustering_Kmeans_Meanshift_HDBSCAN/cleaned_wavelet_radiomics_features.csv"
df = pd.read_csv(input_csv)

# Columns that are NOT features (including ID, folder, and error columns)
non_feature_cols = [
    "PatientID", "FolderType",
    "label_id", "num_voxels", 
    "volume_physical", "YearFolder"

]

# Determine which columns are "actual features" by excluding the columns above
feature_cols = [col for col in df.columns if col not in non_feature_cols]

# Create a boolean column that indicates whether all the feature columns are NaN
df["all_nan_features"] = df[feature_cols].isna().all(axis=1)

# Filter rows where all features are NaN
df_all_nan = df[df["all_nan_features"]]

df_all_nan.head()
# Count how many unique patients in each FolderType have at least one "all-NAN" row
count_unique_patients = df_all_nan.groupby("FolderType")["PatientID"].nunique()

print("Number of unique patients with fully-NaN rows in each FolderType:")
print(count_unique_patients)
```
